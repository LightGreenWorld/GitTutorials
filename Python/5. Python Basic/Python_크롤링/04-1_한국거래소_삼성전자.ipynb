```python
import pandas as pd
from io import BytesIO
from bs4 import BeautifulSoup
import requests
```

```python
## copy > copy link address
url = 'http://marketdata.krx.co.kr/contents/COM/GenerateOTP.jspx'

## copy > copy request headers
headers = {
    'Referer' : 'http://marketdata.krx.co.kr/mdi',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'
}

## copy > copy link address 에서 ? 뒤에 부분
body_data = {
        'name': 'fileDown',
        'filetype': 'xls',
        'url': 'MKD/04/0402/04020100/mkd04020100t3_02',
        'isu_cdnm': 'A005930/삼성전자',
        'isu_cd': 'KR7005930003',
        'isu_nm': '삼성전자',
        'isu_srt_cd': 'A005930',
        'fromdate': '20200518',
        'todate': '20200618',
        'pagePath': '/contents/MKD/04/0402/04020100/MKD04020100T3T2.jsp'
}

r = requests.get(url, body_data, headers=headers)
```

```python
## copy > copy link address
down_url = 'http://file.krx.co.kr/download.jspx'

body_data2 = {
            'code' : r.content
}

r = requests.post(down_url, body_data2, headers = headers)

result = r.content
# print(result)
```

```python
df = pd.read_excel(BytesIO(result))
df.to_csv('Samsung_stock.csv', index=False, index_label=None)
df
```

```python
## 년/월/일 칼럼 list 처리
Date = list(df['년/월/일'])
# Date

## 인덱스와 년/월/일 칼럼 묶도록, 딕셔너리 작업
Date_dict = {}
for i, v in enumerate(Date):
    Date_dict[i] = v
    
Date_dict
```

```python
## 딕셔너리로 묶은 년/월/일 칼럼 인덱스 처리
df2 = df.rename(index=Date_dict)
df2

## df1 = pd.read.csv('Samsung_stock.csv', index_col = '년/월/일')처럼
## 데이터파일 속 칼럼을 바로 인덱스로 지정하면, 칼럼의 title이 인덱스 title로 섷정 
```

```python
## 필요없어진 '년/월/일' 칼럼을 삭제
df2.drop(['년/월/일'], axis = 1, inplace = True)
df2
```

```python
## 데이터프레임 속 문자열 자료를 숫자로 변환
## 1) 콤마 삭제
df2['종가'] = df2['종가'].str.replace(',', '')
df2['대비'] = df2['대비'].str.replace(',', '')
df2['거래량(주)'] = df2['거래량(주)'].str.replace(',', '')
df2['거래대금(원)'] = df2['거래대금(원)'].str.replace(',', '')
df2['시가'] = df2['시가'].str.replace(',', '')
df2['고가'] = df2['고가'].str.replace(',', '')
df2['저가'] = df2['저가'].str.replace(',', '')
df2['시가총액(백만)'] = df2['시가총액(백만)'].str.replace(',', '')
df2['상장주식수(주)'] = df2['상장주식수(주)'].str.replace(',', '')

df2
```

```python
## 2) 숫자 변환 코딩 적용: pd.to_numeric

df2['종가'] = pd.to_numeric(df2['종가'])
df2['대비'] = pd.to_numeric(df2['대비'])
df2['거래량(주)'] = pd.to_numeric(df2['거래량(주)'])
df2['거래대금(원)'] = pd.to_numeric(df2['거래대금(원)'])
df2['시가'] = pd.to_numeric(df2['시가'])
df2['고가'] = pd.to_numeric(df2['고가'])
df2['저가'] = pd.to_numeric(df2['저가'])
df2['시가총액(백만)'] = pd.to_numeric(df2['시가총액(백만)'])
df2['상장주식수(주)'] = pd.to_numeric(df2['상장주식수(주)'])

df2
```
