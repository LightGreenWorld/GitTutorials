### 회귀

```python
import pandas as pd
import matplotlib as mpl
from sklearn.model_selection import train_test_split
```

```python
df3 = pd.read_csv('wave.csv')
x = df3['data'].values.reshape(-1,1)
y = df3['target'].values

X_train, X_test, y_train, y_test = train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model3 = LinearRegression()
model3.fit(X_train, y_train)
```

```python
y_pred = model3.predict(X_test)

import matplotlib.pyplot as pit

pit.scatter(X_test, y_test, label = 'read')
pit.scatter(X_test, y_pred, label = 'pred')
pit.legend
```

```python
df4 = pd.read_csv('boston.csv')

x = df4.drop(['target']), axis1).values
y = df4['target'].values

X_train, X_test, y_train, y_test = train_test_split(x,y)
```

```python
model4 = LinearRegression()
model4.fit(X_train, y_train)
```

```python
y_pred = model4.predict(X_test)

pit.plot(X_test, y_test, label = 'read')
pit.plot(X_test, y_pred, label = 'pred')
pit.legend
```
